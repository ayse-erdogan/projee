```python
filmtablosu=(gelen_veri[0].contents[len(gelen_veri[0].contents)-2]) #contents gelen verinin içindekileri alıyor.
```

```python
print(filmtablosu) #sadece filmlerin olduğu tablo
```

```python
film_tablosu=filmtablosu.find_all("tr") #bağlantıdaki tr leri ekrana bastırır.
#burdaki her bir tr her bir filmi ifade eder.
```

```python
print(film_tablosu)
```

```python
#film başlıklarını ekrana bastır
```

```python
for film in film_tablosu:
    filmbasliklari=film.find_all("td",{"class":"titleColumn"})
    print(filmbasliklari)
    print("******************")
```

```python
#film isimlerini ekrana bastır
for film in film_tablosu:
    filmbasliklari=film.find_all("td",{"class":"titleColumn"})
    film_ismi=filmbasliklari[0].text
    film_ismi=film_ismi.replace("\n","") #her bir "|n" i "" ile değiştir
    print(film_ismi)
    print("******************")
```

```python
#filmlerin imdb puanlarını ekrana bastır
```

```python
for film in film_tablosu:
    poster=film.find_all("td",{"class":"ratingColumn imdbRating"})
    
    print(poster)
    print("******************")
```

```python
for film in film_tablosu:
    puan=film.find_all("td",{"class":"ratingColumn imdbRating"})
    imdbb=puan[0].text
    imdbb=imdbb.replace("\n","") #her bir "|n" i "" ile değiştir
    print(imdbb)
    print("******************")
```

```python
import turtle
import time


screen=turtle.Screen()
screen.setup(width=450,height=450)
screen.bgcolor("yellow")
screen.title("Anolog saat uygulaması")

screen.tracer(0)

kalem=turtle.Turtle()
kalem.speed(0)
kalem.pensize(3)


def cizim(saat,dakika,saniye,kalem):
    kalem.begin_fill()
    kalem.color("black")
    kalem.up()
    
    kalem.goto(0,210)
    kalem.setheading(180)
    kalem.color("brown")
    kalem.pendown()
    kalem.circle(210)
    kalem.end_fill()
    
    kalem.up()
    kalem.goto(0,0)
    kalem.setheading(90)
    kalem.begin_fill()
    kalem.color("white")
    
    
    #12 tane çizgi çiziyor.
    for i in range(12):
        kalem.fd(190)
        kalem.pendown()
        kalem.fd(20)
        kalem.up()
        kalem.goto(0,0)
        kalem.rt(30)
        
        
    kalem.end_fill()
    

    
    #saniye
    
    kalem.up()
    kalem.goto(0,0)
    kalem.color("red")
    kalem.setheading(90)
    bilgi=(saniye/60)*360
    kalem.rt(bilgi)
    kalem.pendown()
    kalem.fd(170)   
    
    
    #dakika
    kalem.up()
    kalem.goto(0,0)
    kalem.color("black")
    kalem.setheading(90)
    bilgi=(dakika/60)*360
    kalem.rt(bilgi)
    kalem.pendown()
    kalem.fd(140)
    
    
    #saat
    kalem.up()
    kalem.goto(0,0)
    kalem.color("blue")
    kalem.setheading(90)
    bilgi=(saat/12)*360
    kalem.rt(bilgi)#açıları belirtiyoruz açıya göre dönüyor
    kalem.pendown()
    kalem.fd(100)
    

#saatin dönmesi için oluşturduğum döngü
while True:    
    saat=int(time.strftime("%I"))  
    dakika=int(time.strftime("%M"))
    saniye=int(time.strftime("%S"))
    cizim(saat,dakika,saniye,kalem)
    screen.update() #ekranı güncelle
    kalem.clear()
```
